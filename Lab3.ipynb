### Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

````python
# Definicion de arrays
a = np.array([1, 4, 2, 7])
b = np.array([1, 3, 2, 9])

# calculamos la diferencia entre los arrays a y b
diferencia = a - b

# Calcular la media aritmética de la diferencia
m_aritmetica = np.mean(diferencia)

print("La media aritmetica es:", m_aritmetica)
````

### 5) Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

````python
granada = dataset[dataset['Provincia'] == 'Granada']

# Calcular la media, mediana, desviación estándar, valor máximo y valor mínimo de la población
media_poblacion = granada['Poblacion'].mean()
mediana_poblacion = granada['Poblacion'].median()
desviacion_estandar_poblacion = granada['Poblacion'].std()
maximo_poblacion = granada['Poblacion'].max()
minimo_poblacion = granada['Poblacion'].min()

print("Media de la población:", media_poblacion)
print("Mediana de la población:", mediana_poblacion)
print("Desviación estándar de la población:", desviacion_estandar_poblacion)
print("Poblacion maxima:", maximo_poblacion)
print("Poblacion minima:", minimo_poblacion)
````

### 8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

````python
dataset.head()
madrid_data = dataset[dataset['Municipio'] == 'Madrid']
madrid_data

madrid_coords = (madrid_data['X'].values[0], madrid_data['Y'].values[0])

print("Coordenadas de Madrid:", madrid_coords)
````

````python
def distancia_entre_coords(coords1, coords2):
    lat1, lon1 = coords1
    lat2, lon2 = coords2
    distancia = np.sqrt((lat2 - lat1)**2 + (lon2 - lon1)**2)
    return distancia

distancias_madrid = []
for index, municipio in dataset.iterrows():
    coords_municipio = (municipio['X'], municipio['Y'])
    distancia = distancia_entre_coords(madrid_coords, coords_municipio)
    if (municipio['Municipio'] != 'Madrid'):
        distancias_madrid.append((municipio['Municipio'], distancia))

distancias_madrid

# Encontrar el municipio más cercano y el más lejano a Madrid
municipio_mas_cercano = min(distancias_madrid, key=lambda x: x[1])
municipio_mas_lejano = max(distancias_madrid, key=lambda x: x[1])

print("Municipio más cercano a Madrid:", municipio_mas_cercano[0], "- Distancia:", municipio_mas_cercano[1])
print("Municipio más lejano a Madrid:", municipio_mas_lejano[0], "- Distancia:", municipio_mas_lejano[1])
````
